```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data
```

```python
from env import get_db_url
```

```python
url = get_db_url("iris_db")

meas_table = """
SELECT * 
FROM measurements"""

measurements = pd.read_sql(meas_table, url)
measurements.head()
```

```python
sns.histplot(measurements.petal_length, kde = True)
plt.show()
```

```python
sns.relplot(x=measurements.petal_width, y= measurements.petal_length)
plt.title('Petal Length and Width Correlation')
```

```python
sns.relplot(x=measurements.species_id, y= measurements.sepal_length)
```

```python
sns.relplot(x=measurements.species_id, y= measurements.sepal_width)
```
